```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['svg.fonttype'] = 'none'
from scipy import stats
```

```python
penetrating_coactivation = pd.read_csv("penetrating\\coactivation_norm.csv",index_col=0)
epineural_coactivation = pd.read_csv("epineural\\coactivation_norm.csv",index_col=0)
```

```python
asd = penetrating_coactivation.values.tolist()

penetrating = []
for i in asd:
    penetrating.extend(i)
    
penetrating
```

```python
qwe = epineural_coactivation.values.tolist()

epineural = []
for i in qwe:
    epineural.extend(i)
    
epineural
```

```python
plt.subplots(figsize = (7,7))
fig = sns.regplot(penetrating, epineural,ci=None, x_jitter = 0.05, y_jitter = 0.05)
# fig = sns.regplot(penetrating, epineural,ci=None)

slope, intercept, r_value, p_value, std_err = stats.linregress(penetrating, epineural)
print slope, intercept, r_value, p_value, std_err

r_val = stats.pearsonr(penetrating, epineural)
print r_val

# plt.savefig('regress_epi_pen_coactivation.svg')
```
